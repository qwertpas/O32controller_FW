```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib widget

input_file = "./txt/rpmlog.txt"
output_file = "./csv/rpmlog.csv"

data = []
with open(input_file, 'r') as file:
    lines = file.readlines()
    for i in range(0, len(lines)-2, 2):
        column1 = lines[i].strip()
        column2 = lines[i+1].strip()
        data.append([column1, column2])

with open(output_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['0', '1'])
    writer.writerows(data)

df = pd.read_csv(output_file)
df['rpm'] = (df['0']*2**7) + df['1']

def twos_complement_to_signed(value, num_bits=14):
    if value & (1 << (num_bits - 1)):
        value = value - (1 << num_bits)
    return value

df['rpm'] = df['rpm'].apply(twos_complement_to_signed)


# df.plot(y=['0', '1'], style='-')
df.plot(y=['rpm'], style='-')
```

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib widget

input_file = "./txt/spotty.txt"
output_file = "./csv/spotty.csv"

data = []
with open(input_file, 'r') as file:
    lines = file.readlines()
    for i in range(0, len(lines)-2, 3):
        column1 = lines[i].strip()
        column2 = lines[i+1].strip()
        column3 = lines[i+2].strip()
        data.append([column1, column2, column3])

with open(output_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['cmd', 'data1', 'data0'])
    writer.writerows(data)

df = pd.read_csv(output_file)
df['data14'] = (df['data1']*2**7) + df['data0']
df['data14'] = df['data14'] / 2**7

df.plot(y=['cmd', 'data1', 'data0'])
```

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib widget

input_file = "./txt/lessspotty.txt"
output_file = "./csv/lessspotty.csv"

data = []
with open(input_file, 'r') as file:
    lines = file.readlines()
    for i in range(0, len(lines)-2, 3):
        column1 = lines[i].strip()
        column2 = lines[i+1].strip()
        column3 = lines[i+2].strip()
        data.append([column1, column2, column3])

with open(output_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['cmd', 'data1', 'data0'])
    writer.writerows(data)

df = pd.read_csv(output_file)
df['data14'] = df['data1'].apply(lambda x: x << 7) | df['data0']
df['data14'] = df['data14']

df.plot(y=['cmd', 'data0', 'data1'])
```
